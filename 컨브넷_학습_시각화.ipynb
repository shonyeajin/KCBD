```python
img_path='./datasets/cats_and_dogs_small/test/cats/cat.1700.jpg'

from keras.preprocessing import image
import numpy as np

img=image.load_img(img_path, target_size=(150,150))
img_tensor=image.img_to_array(img)
img_tensor=np.expand_dims(img_tensor,axis=0)
img_tensor/=255.

print(img_tensor.shape)
```

```python
import matplotlib.pyplot as plt

plt.imshow(img_tensor[0])
plt.show()
```

***중간층의 활성화 시각화 하기***

```python
from keras import models

layer_outputs= [layer.output for layer in model.layers[:8]]
layer_outputs
```

```python
activation_model=models.Model(inputs=model.input, outputs=layer_outputs)

activation_model.summary()
```

load해온 모델의 상위 8개 층을 추출하여 입력에 대해 8개 층 각각의 층의 출력을 반환하는 모델을 만듬

```python
activations=activation_model.predict(img_tensor)
```

```python
first_layer_activation=activations[0]
print(first_layer_activation.shape)
```

당연히 위의 summary와 일치.<br>
32개의 채널을 갖는 148X148 크기의 특성 맵이다.

```python
import matplotlib.pyplot as plt
plt.matshow(first_layer_activation[0,:,:,6], cmap='viridis')
```

중간층의 모든 활성화에 있는 채널 시각화 하기

```python
layer_names=[]
for layer in model.layers[:8]:
  layer_names.append(layer.name)

images_per_row=16


for layer_name, layer_activation in zip(layer_names, activations):
  n_features=layer_activation.shape[-1] #각 레이어의 출력의 필터의 갯수
  
  size=layer_activation.shape[1]

  n_cols=n_features//images_per_row
  display_grid=np.zeros((size*n_cols,images_per_row*size))

  for col in range(n_cols):
    for row in range(images_per_row):
      channel_image=layer_activation[0,:,:,col*images_per_row+row]
      channel_image-=channel_image.mean()
      channel_image/=channel_image.std()
      channel_image*=64
      channel_image+=128
      channel_image=np.clip(channel_image, 0,255).astype('uint8')
      display_grid[col*size:(col+1)*size, row*size:(row+1)*size]=channel_image
  scale=1./size
  plt.figure(figsize=(scale*display_grid.shape[1], scale* display_grid.shape[0]))
  plt.title(layer_name)
  plt.grid(False)
  plt.imshow(display_grid, aspect='auto', cmap='viridis')
plt.show()
```
